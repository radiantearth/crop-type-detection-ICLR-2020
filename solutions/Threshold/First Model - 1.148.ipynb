```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from tqdm import tqdm as tqdm
```

```python
train = pd.read_csv("train ICLR.csv")
```

```python
train = train.drop(["Unnamed: 0"], axis=1)
```

```python
train.head()
```

```python
X = train[train.columns.difference(['Field_ID', 'label'])]
```

```python
X
```

```python
y = train['label']
```

```python
import catboost as ctb
import lightgbm as ltb
import xgboost as xgb
```

```python
ctb1 = ctb.CatBoostClassifier(iterations=1700)
```

```python
model=ctb1.fit(X,y)
```

```python
test = pd.read_csv("test ICLR.csv")
```

```python
test
```

```python
test = test.drop(["Unnamed: 0"], axis=1)
```

```python
X_test = test[test.columns.difference(['Field_ID', 'label'])]
```

```python
X_test.head()
```

```python
y_pred = model.predict_proba(X_test)
```

```python
y_pred
```

```python
prediction = pd.DataFrame(y_pred, columns=['Crop_ID_1', 'Crop_ID_2', 'Crop_ID_3', 'Crop_ID_4', 'Crop_ID_5', 'Crop_ID_6', 'Crop_ID_7'])
```

```python
test
```

```python
test = test.reset_index()
```

```python
submission = pd.concat([test['Field_ID'], prediction], axis=1)
```

```python
submission.head()
```

```python
for i in tqdm(range(len(submission))):
    for j in range(1,8):
        if submission.iloc[i][j] > 0.90:
            submission.loc[i, submission.columns.difference([submission.columns[0]])] = 0
            submission.loc[i, submission.columns[j]] = 1
```

```python
for i in tqdm(range(len(submission))):
    for j in range(1,8):
        if submission.iloc[i][j] > 0.85:
            #submission.loc[i, submission.columns.difference([submission.columns[0]])] = 0
            submission.loc[i, submission.columns[j]] = 1
```

```python
for i in tqdm(range(len(submission))):
    for j in range(5,8):
        if submission.iloc[i][j] == max(submission.loc[i, submission.columns.difference([submission.columns[0]])]):
            submission.loc[i, submission.columns.difference([submission.columns[0]])] = 0
            submission.loc[i, submission.columns[j]] = 1
```
